```python
#Create train test split 

from sklearn.model_selection import train_test_split
X = df.drop('Final Mark', axis =1)
y = df.loc[:,['Final Mark']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Model BaseLine

```python
#createModel Baseline

#import cross validation score
from sklearn.model_selection import cross_val_score

#import DummyRegressor 
from sklearn.dummy import DummyRegressor


#create classifier object
dr = DummyRegressor()

#run cv for NB classifier
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error



dr_mae = cross_val_score(dr,X_train,y_train.values.ravel(), cv=3, scoring ='neg_mean_absolute_error')
dr_mse = cross_val_score(dr,X_train,y_train.values.ravel(), cv=3, scoring ='neg_mean_squared_error')

print('dr_mae: ' +str(dr_mae))
print('dr_mse : '+str(dr_mse))
print('dr_mae_avg: ' + str(dr_mae.mean()) +'  |  lr_mse_avg: '+str(dr_mse.mean()))

```

In this section we will try:

- Iinear regression
- SVR
- XGBoost

```python
#Let's now experiment with a few different basic models 

## LinearRegression
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr_mae = cross_val_score(lr,X_train,y_train.values.ravel(), cv=3, scoring ='neg_mean_absolute_error')
lr_mse = cross_val_score(lr,X_train,y_train.values.ravel(), cv=3, scoring ='neg_mean_squared_error')

print('lr_mae: ' +str(lr_mae))
print('lr_mse: '+str(lr_mse))
print('lr_mae_avg: ' + str(lr_mae.mean()) +'  |  lr_mse_avg: '+str(lr_mse.mean())+'\n')

##
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline, Pipeline

lr_scaling = make_pipeline(StandardScaler(), LinearRegression())

lr_mae_scaling = cross_val_score(lr_scaling,X_train,y_train.values.ravel(), cv=3, scoring ='neg_mean_absolute_error')
lr_mse_scaling = cross_val_score(lr_scaling,X_train,y_train.values.ravel(), cv=3, scoring ='neg_mean_squared_error')

print('lr_mae_scaling: ' +str(lr_mae_scaling))
print('lr_mse_scaling: '+str(lr_mse_scaling))
print('lr_mae_scaling_avg: ' + str(lr_mae_scaling.mean()) +'  |  lr_mse_scaling_avg: '+str(lr_mse_scaling.mean())+'\n')



## SVR
from sklearn.svm import SVR
svr = SVR()

svr_mae = cross_val_score(svr,X_train,y_train.values.ravel(), cv=3, scoring ='neg_mean_absolute_error')
svr_mse = cross_val_score(svr,X_train,y_train.values.ravel(), cv=3, scoring ='neg_mean_squared_error')

print('svr_mae: ' +str(svr_mae))
print('svr_mse : '+str(svr_mse))
print('svr_mae_avg: ' + str(svr_mae.mean()) +'  |  svr_mse_avg: '+str(svr_mse.mean())+'\n')


## XGBoost 
from xgboost.sklearn import XGBRegressor

xgboost = XGBRegressor()
xg_mae = cross_val_score(xgboost,X_train,y_train.values.ravel(), cv=3, scoring ='neg_mean_absolute_error')
xg_mse = cross_val_score(xgboost,X_train,y_train.values.ravel(), cv=3, scoring ='neg_mean_squared_error')

print('xg_mae: ' +str(xg_mae))
print('xg_mse : '+str(xg_mse))
print('xg_mae_avg: ' + str(xg_mae.mean()) +'  |  xg_mse_avg: '+str(xg_mse.mean())+'\n')
```

```python
##Model Comparison
data = [    [str(dr_mae.mean()*-1),str(dr_mse.mean()*-1)],
            [str(lr_mae.mean()*-1), str(lr_mse.mean()*-1)],
            [str(svr_mae.mean()*-1), str(svr_mse.mean()*-1)],
            [str(xg_mae.mean()*-1),str( xg_mse.mean()*-1)]]

df_comparison = pd.DataFrame(data, columns=['MAE','MSE'] , index=['DummyRegressor','LinearRegression','SVR', 'XGBRegressor'])
df_comparison
```

So , the best model so far is Linear Regressor, then XGB.

Let's try to Tune Paramter of SVR

```python
##RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV

svr = SVR()

param_grid = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001],'kernel': ['rbf', 'poly', 'sigmoid']}


rs_svr = RandomizedSearchCV(estimator = svr, param_distributions =param_grid, n_iter =100, cv = 3, random_state = 42, scoring ='neg_mean_absolute_error')

rs_svr.fit(X_train,y_train)
```

```python
## GridSearchCV
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001],'kernel': ['rbf', 'poly', 'sigmoid']}
gs_svr = GridSearchCV(estimator = svr, param_grid =param_grid,  cv = 3,  scoring ='neg_mean_absolute_error')
gs_svr.fit(X_train,y_train)
```
